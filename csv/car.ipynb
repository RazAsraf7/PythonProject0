```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('2024_car_buy.csv')

# Grouping by 'make' and counting the occurrences
model_group = df.groupby('make').count()
print(model_group)

# Sorting the DataFrame by the count of cars ('year' column)
model_group_sorted = model_group.sort_values(by='year', ascending=True)

# Plotting
fig, ax = plt.subplots(figsize=(10, 6))
ax.barh(model_group_sorted.index, model_group_sorted['year'],height=0.5)


# Add annotations for survived individuals
for bar, count in zip(model_group_sorted.index, model_group_sorted['year']):
    ax.text(count, bar, str(count), va='center')

# Adding labels and title
ax.set_ylabel('Brand Of Car', rotation=0,labelpad=40)
ax.set_title('The most buying brand car in 2024')

# Remove right and top spines
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.spines['bottom'].set_visible(False)

# Remove ticks on x-axis
ax.xaxis.set_ticks([])


# Increase spacing between y-axis tick labels
ax.set_yticklabels(model_group_sorted.index, fontsize=8)  # Adjust font size
ax.tick_params(axis='y', which='major', pad=5)  # Increase padding between tick labels

plt.show()

```

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt

df=pd.read_csv('data_car.csv')

# print(df)

# Find the model with the highest weight
model_with_highest_weight = df.loc[df['Weight'].idxmax(), 'Model']
# Find the model with the highest weight
model_with_lowest_weight = df.loc[df['Weight'].idxmin(), 'Model']

list_model=[model_with_highest_weight,model_with_lowest_weight]
list_wieght=[int(min(df['Weight'])),int(max(df['Weight']))]


print(list_wieght)
# Create a scatter plot
fig, ax = plt.subplots()
ax.scatter(list_model, list_wieght, alpha=0.5)

 # Set title and axis labels
ax.set_title("Model with highest weight and lowest weight")
ax.set_xlabel("Model")
# Rotate Y-axis labels
ax.set_ylabel(
'Weight', 
rotation=0, 
ha='right', 
ma='left'
) 

ax.set_yticks([int(min(df['Weight'])), int(max(df['Weight']))])

# Remove right and top spines
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)

plt.show()

```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming df_2024_car_buy is your DataFrame containing the 2024 car purchase data.
# Count the number of models for each car make and sort them in descending order for the plot
car_make_counts = df_2024_car_buy['make'].value_counts().sort_values(ascending=True)

# Create the horizontal bar plot
plt.figure(figsize=(10, 8))  # Increased figure size for better visibility
ax = plt.gca()  # Get current axis

# Adding the horizontal bars
bars = ax.barh(car_make_counts.index, car_make_counts, color='skyblue', height=0.5)  # Adjust bar height for spacing

# Add labels to the bars
for bar in bars:
    width = bar.get_width()
    label_x_pos = width + 1  # Offset the label a little bit to the right
    ax.text(label_x_pos, bar.get_y() + bar.get_height()/2, f'{width}', va='center')

# Title and labels
plt.title('Market Share by Car Make for 2024 Models')
plt.xlabel('Number of Models')
plt.ylabel('Car Make')

# Remove the right and left spines
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(False)

# Set rotation for y-axis labels
plt.yticks(rotation=0)

# Show the plot
plt.show()
```
